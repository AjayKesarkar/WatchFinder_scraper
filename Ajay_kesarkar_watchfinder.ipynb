### Importing Libraries

```python
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver = webdriver.Chrome(executable_path='C:\webdriver\chromedriver.exe')
driver.get('https://www.watchfinder.co.uk/new-arrivals')
driver.maximize_window()
time.sleep(3)

# Clicking the "Continue" button
continue_button = driver.find_element_by_xpath('//*[@id="modal_region-selector"]/div/div/div[1]/button')
continue_button.click()

# Clicking the "Accept All" button for cookies
accept_all = driver.find_element_by_xpath('//*[@id="cookie_"]/div/div/div[2]/div[2]/button/span')
accept_all.click()

# Lists to store watch details
Img = []
Name = []
Price = []
Currency = []
Model = []
Model_number = []
Year = []
Box = []
Papers = []
Product_code = []
Warranty = []
Case_size = []
Case_material = []
Movement = []
Bracelet_Material = []
Dial_Type = []
Water_Resistance = []
Location = []
Description = []

img_elements = driver.find_elements_by_xpath("//div[@class='relative flex msm:h-60']")
# img_elements = driver.find_elements_by_css_selector('[data-testid="watchItem"]')
for i in range(len(img_elements)):
    # Re-locate the image elements within the loop
    img_elements = driver.find_elements_by_xpath("//div[@class='relative flex msm:h-60']")
#     img_elements = driver.find_elements_by_css_selector('[data-testid="watchItem"]')
    img_elements[i].click()

    # Brand Name
    brand_name = driver.find_element_by_xpath('//*[@id="content"]/section[3]/div[1]/h1/span[1]').text
    Name.append(brand_name if brand_name else 'N/A')

    # Price
    price = driver.find_element_by_xpath('//*[@id="content"]/section[3]/div[1]/span[1]').text
    Price.append(price if price else 'N/A')
    Currency.append(price[0] if price else 'N/A')

    # Model
    model = driver.find_element_by_xpath('//*[@id="content"]/section[3]/div[1]/h1/span[2]').text
    Model.append(model if model else 'N/A')

    # Model number
    model_number = driver.find_element_by_xpath('//*[@id="content"]/section[3]/div[1]/h1/span[3]').text
    Model_number.append(model_number if model_number else 'N/A')

    # Year
    year = driver.find_element_by_xpath('//*[@id="specification-content"]/table/tbody/tr[3]/td[2]').text
    Year.append(year if year else 'N/A')

    # Box
    box = driver.find_element_by_xpath('//*[@id="specification-content"]/table/tbody/tr[1]/td[2]').text
    Box.append(box if box else 'N/A')

    # Papers
    papers = driver.find_element_by_xpath('//*[@id="specification-content"]/table/tbody/tr[2]/td[2]').text
    Papers.append(papers if papers else 'N/A')

    # Product code
    product_code = driver.find_element_by_xpath('//*[@id="specification-content"]/table/tbody/tr[4]/td[2]').text
    Product_code.append(product_code if product_code else 'N/A')
    
    # Warranty 
    
    warranty = driver.find_elements_by_xpath('//*[@id="specification-content"]/table/tbody/tr[5]/td[2]')
    for w in warranty:
        if w is None:
            Warranty.append('N/A')
        else:
            Warranty.append(w.text)

    # Case size 
    cs = driver.find_elements_by_xpath('//*[@id="specification-content"]/table/tbody/tr[6]/td[2]')
    for c in cs:
        if c is None:
            Case_size.append('N/A')
        else:
            Case_size.append(c.text)


    # Case material 
    cm = driver.find_elements_by_xpath('//*[@id="specification-content"]/table/tbody/tr[7]/td[2]')
    for c in cm:
        if c is None:
            Case_material.append('N/A')
        else:
            Case_material.append(c.text)


    # Movement 
    movement = driver.find_elements_by_xpath('//*[@id="specification-content"]/table/tbody/tr[8]/td[2]')
    for m in movement:
        if m is None:
            Movement.append('N/A')
        else:
            Movement.append(m.text)


    # Bracelet Material 
    bm = driver.find_elements_by_xpath('//*[@id="specification-content"]/table/tbody/tr[9]/td[2]')
    for b in bm:
        if b is None:
            Bracelet_Material.append('N/A')
        else:
            Bracelet_Material.append(b.text)

    # Dial Type
    dt = driver.find_elements_by_xpath('//*[@id="specification-content"]/table/tbody/tr[10]/td[2]')
    for d in dt:
        if d is None:
            Dial_Type.append('N/A')
        else:
            Dial_Type.append(d.text)

    # Water Resistance
    wr = driver.find_elements_by_xpath('//*[@id="specification-content"]/table/tbody/tr[11]/td[2]')
    for w in wr:
        if w is None:
            Water_Resistance.append('N/A')
        else:
            Water_Resistance.append(w.text)

    # Location
    location = driver.find_elements_by_xpath('//*[@id="specification-content"]/table/tbody/tr[12]/td[2]')
    for l in location:
        if l is None:
            Location.append('N/A')
        else:
            Location.append(l.text)

    # Description
    description = driver.find_elements_by_xpath('//*[@id="description-content"]/p')
    for d in description:
        if d is None:
            Description.append('N/A')
        else:
            Description.append(d.text)
    
    time.sleep(3)
    driver.back()
    
    driver.refresh()
    time.sleep(3)
    
print(Name)
```
